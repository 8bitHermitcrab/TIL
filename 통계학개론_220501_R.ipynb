# 문제2.

```r
# 학생 10명이 1분당 할 수 있는 윗몸일으키기 개수
students2 <- c(25, 41, 35, 8, 52, 23, 32, 37, 42, 28)
```

```r
t.test(students2)
```

```r
# z-test (모집단의 분산을 아는 경우)

# 표본평균
m = mean(students2)
print(m)

# 검정통계량 Z 계산
z = (m - 30) / 5 * sqrt(10)
print(z)

# 유의확률(p-value)
p = (1 - pnorm(z)) * 2
print(p)
```

# 문제6.

```r
# 학생 번호
students_nums <- c(1, 2, 3, 4, 5, 6, 7)
# 방과 후 학습 전
before_scores <- c(72, 80, 83, 63, 66, 76, 82)
# 방과 후 학습 후
after_scores <- c(78, 82, 82, 68, 70, 75, 88)

data6 <- data.frame(students_nums, before_scores, after_scores)
data6
```

```r
# 산점도
plot(formula=after_scores~before_scores, data=data6,
    main='Scatter plot',
    xlab='before_scores',
    ylab='after_scores')
```

```r
# 상관계수
cor(before_scores, after_scores)
```

```r
# 회귀직선
# 독립변수 : 방과 후 학습 전 성취도 점수
# 종속변수 : 방과 후 학습 후 성취도 점수
reg <- lm(after_scores~before_scores)
summary(reg)
plot(before_scores, after_scores)
abline(reg)
```
